```python
def matching_valuation(x):
    # print(x)
    standard_ratio = 10/11
    if x['Valuation Grade'] == "A+":
        return 11*standard_ratio
    elif x['Valuation Grade'] == "A":
        return 10*standard_ratio
    elif x['Valuation Grade'] == "A-":
        return 9*standard_ratio
    elif x['Valuation Grade'] == "B+":
        return 8*standard_ratio
    elif x['Valuation Grade'] == "B":
        return 7*standard_ratio
    elif x['Valuation Grade'] == "B-":
        return 6*standard_ratio
    elif x['Valuation Grade'] == "C+":
        return 5*standard_ratio
    elif x['Valuation Grade'] == "C":
        return 4*standard_ratio
    elif x['Valuation Grade'] == "C-":
        return 3*standard_ratio
    elif x['Valuation Grade'] == "D+":
        return 2*standard_ratio
    elif x['Valuation Grade'] == "D":
        return 1*standard_ratio

def matching_profitability(x):
    standard_ratio = 10/13
    if x['Profitability Grade'] == "A+":
        return 13*standard_ratio
    elif x['Profitability Grade'] == "A":
        return 12*standard_ratio
    elif x['Profitability Grade'] == "A-":
        return 11*standard_ratio
    elif x['Profitability Grade'] == "B+":
        return 10*standard_ratio
    elif x['Profitability Grade'] == "B":
        return 9*standard_ratio
    elif x['Profitability Grade'] == "B-":
        return 8*standard_ratio
    elif x['Profitability Grade'] == "C+":
        return 7*standard_ratio
    elif x['Profitability Grade'] == "C":
        return 6*standard_ratio
    elif x['Profitability Grade'] == "C-":
        return 5*standard_ratio
    elif x['Profitability Grade'] == "D+":
        return 4*standard_ratio
    elif x['Profitability Grade'] == "D":
        return 3*standard_ratio
    elif x['Profitability Grade'] == "D-":
        return 2*standard_ratio
    elif x['Profitability Grade'] == "F":
        return 1*standard_ratio
    
def matching_growth(x):
    standard_ratio = 10/13
    if x['Growth Grade'] == "A+":
        return 13*standard_ratio
    elif x['Growth Grade'] == "A":
        return 12*standard_ratio
    elif x['Growth Grade'] == "A-":
        return 11*standard_ratio
    elif x['Growth Grade'] == "B+":
        return 10*standard_ratio
    elif x['Growth Grade'] == "B":
        return 9*standard_ratio
    elif x['Growth Grade'] == "B-":
        return 8*standard_ratio
    elif x['Growth Grade'] == "C+":
        return 7*standard_ratio
    elif x['Growth Grade'] == "C":
        return 6*standard_ratio
    elif x['Growth Grade'] == "C-":
        return 5*standard_ratio
    elif x['Growth Grade'] == "D+":
        return 4*standard_ratio
    elif x['Growth Grade'] == "D":
        return 3*standard_ratio
    elif x['Growth Grade'] == "D-":
        return 2*standard_ratio
    elif x['Growth Grade'] == "F":
        return 1*standard_ratio
    

def matching_price_performance(x):
    standard_ratio = 10/13
    if x['Performance Grade'] == "A+":
        return 13*standard_ratio
    elif x['Performance Grade'] == "A":
        return 12*standard_ratio
    elif x['Performance Grade'] == "A-":
        return 11*standard_ratio
    elif x['Performance Grade'] == "B+":
        return 10*standard_ratio
    elif x['Performance Grade'] == "B":
        return 9*standard_ratio
    elif x['Performance Grade'] == "B-":
        return 8*standard_ratio
    elif x['Performance Grade'] == "C+":
        return 7*standard_ratio
    elif x['Performance Grade'] == "C":
        return 6*standard_ratio
    elif x['Performance Grade'] == "C-":
        return 5*standard_ratio
    elif x['Performance Grade'] == "D+":
        return 4*standard_ratio
    elif x['Performance Grade'] == "D":
        return 3*standard_ratio
    elif x['Performance Grade'] == "D-":
        return 2*standard_ratio
    elif x['Performance Grade'] == "F":
        return 1*standard_ratio
```

```python
data['Valuation Grade']
```

```python
data['Valuation Score'] = data.apply(lambda x: matching_valuation(x), axis = 1)
data['Profitability Score'] = data.apply(lambda x: matching_profitability(x), axis = 1)
data['Growth Score'] = data.apply(lambda x: matching_growth(x), axis = 1)
data['PricePerformance Score'] = data.apply(lambda x: matching_price_performance(x), axis = 1)
```

```python
data
```

```python
data.sort_values("Overall Rating", ascending=False)
```

```python
ticker = "ASC"
val = round(data.loc[data['Ticker']==ticker]['Valuation Score'].values[0], 2)
prof = round(data.loc[data['Ticker']==ticker]['Profitability Score'].values[0], 2)
growth = round(data.loc[data['Ticker']==ticker]['Growth Score'].values[0], 2)
price = round(data.loc[data['Ticker']==ticker]['PricePerformance Score'].values[0], 2)
# imagine you will have this sentiment score below that is calculated from another code source
sentiment = 9.16
```

```python
company_name = data.loc[data['Ticker']==ticker]['Company'].values[0]
mkt_cap = data.loc[data['Ticker']==ticker]['Market Cap'].values[0]
sector = data.loc[data['Ticker']==ticker]['Sector'].values[0]
country = data.loc[data['Ticker']==ticker]['Country'].values[0]
overall_rating = data.loc[data['Ticker']==ticker]['Overall Rating'].values[0]
```

```python
overall_rating
```

```python
fig = go.Figure(data=go.Scatterpolar(
  r=[val, prof, growth, price, sentiment],
  theta=[f'Valuation Score: {val}',f'Profitability Score: {prof}',
         f'Growth Score: {growth}', f'Price Momentum Score: {price}',f'Sentiment Score: {sentiment}'],
  fill='toself',
  name=f"Company Ticker: {ticker}"
))

fig.update_layout(
  polar=dict(
    radialaxis=dict(
      visible=True,
      range=[0,10],
      tickfont=dict(size=20))
  ),
  width=1200,
  height=600,
  showlegend=True,
  font=dict(size=20)
)
print(f"Company Name: {company_name}")
print(f"Market Cap: {mkt_cap}")
print(f"Sector: {sector}")
print(f"Country: {country}")
print(f"Overall Rating: {overall_rating}")

fig.show()
```
